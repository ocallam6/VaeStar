# Introduction

In this notebook we are going to probabilitsitcally find the distribution away from the fitted isochrone. Our first assumption is that the isochrone is properly fit. Therefore, no algorithmic fitting has been applied to the isochrone via our data or its error.

The first thing to note is that we will define a point $\vec{\hat{x}}$ in the HR diagram as having distribution $$ \vec{\hat{x}} \sim N(\vec{x},\Sigma) $$
where the hat denotes that it is an observed value.

Now we can marginalse out each of the observations
$$ \hat x_g \sim N(x_g,\sigma^2_g) $$
$$ \hat x_c \sim N(x_c,\sigma^2_c) $$
so that the first corresponds to the observation of the absolute magnitude and the second denotes the Gaia colour. We will assume their measurement is independent of eachother.

For a given point, let $Ax+By+c=0$ be the equation of the line of best fit closest to a point of interest perpendicular locally to the isochrone. Our tranformation will be $$ y=(Bx_g+Ax_c+c)\over \sqrt{A^2+B^2}$$

Which, from the properties of Normal Distributions, has a distribution $$\hat y \sim N((Bx_g+Ax_c+c)/\sqrt{A^2+B^2},(B^2 \sigma^2_g+A^2 \sigma^2_c)/ \sqrt{A^2+B^2})$$

If we assume that the values $y=(Bx_g+Ax_c+c)/\sqrt{A^2+B^2}$ can be modelled with a Gaussian Mixture Model $$p(y)=\sum_j \alpha_j N(\bar y|m_j,\sigma^2_j)$$

Then using propetries of Gaussians we have that the observed $\hat y$ has the distribution $$p(\hat y)=\sum_j \alpha_j N(\bar y|m_j,\sigma^2_j+(B^2 \sigma^2_g+A^2 \sigma^2_c)/ \sqrt{A^2+B^2}),$$
which is also a Gaussian Mixture.

Actually we make another transformation by dividing by 1/sqrt(1+pslopes squared) to project noto the difference of colour mag

```python
p_slopes=[]

x=isochrone['BPRP']
y=isochrone['Gaia_G_DR2Rev']
for i in range(len(isochrone)-1):    
    dy=y[i+1]-y[i]
    dx=x[i+1]-x[i]
    p_slopes.append(-1*dx/dy)
p_slopes.append(0)
```

```python
isochrone['p_slopes']=p_slopes
isochrone['slopes']=-1/isochrone['p_slopes']

high_c=[]
low_c=[]
for i in range(len(isochrone)-1):
    high_c.append(isochrone['Gaia_G_DR2Rev'][i+1] - isochrone['p_slopes'][i]*isochrone['BPRP'][i+1])
    low_c.append(isochrone['Gaia_G_DR2Rev'][i] - isochrone['p_slopes'][i]*isochrone['BPRP'][i])
high_c.append(0)
low_c.append(0)
```

Set the Bounds of selection region for each part of the isochrone

```python
isochrone['low_c']=high_c #high c is low c oops
isochrone['high_c']=low_c
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler(with_mean=False,with_std=False)
#should it be true or false on the mean
```

```python
isochrone=isochrone.loc()[0:len(isochrone)-2]
centers=[]
tests=[]
err_means=[]
for i in range(35):
    try:
        x=(orig_scatter['mean_absolute_mag_g_band']-orig_scatter['bp_rp']*isochrone['p_slopes'][i]<=isochrone['high_c'][i])
        y=(isochrone['low_c'][i]<=orig_scatter['mean_absolute_mag_g_band']-orig_scatter['bp_rp']*isochrone['p_slopes'][i])
        
        test=orig_scatter[x]
        test=test[y]
        
        #new method

        center=isochrone['Gaia_G_DR2Rev'][i] - isochrone['slopes'][i]*isochrone['BPRP'][i]
        test['yproj']=((1/np.sqrt(1+isochrone['slopes'][i]**2))*(test['mean_absolute_mag_g_band']-test['bp_rp']*isochrone['slopes'][i]-isochrone['Gaia_G_DR2Rev'][i] + isochrone['slopes'][i]*isochrone['BPRP'][i]))*(1/np.sqrt(1+isochrone['p_slopes'][i]**2))
        test['yproj_err']=((1/(1+isochrone['slopes'][i]**2))*test['phot_g_mean_mag_error']**2+(isochrone['slopes'][i]*test['bp_rp_error'])**2)*(1/(1+isochrone['p_slopes'][i]**2))

        test['y']=((1/np.sqrt(1+isochrone['slopes'][i]**2))*(test['mean_absolute_mag_g_band']-test['bp_rp']*isochrone['slopes'][i]-isochrone['Gaia_G_DR2Rev'][i] + isochrone['slopes'][i]*isochrone['BPRP'][i]))
        test['y_err']=((1/(1+isochrone['slopes'][i]**2))*test['phot_g_mean_mag_error']**2+(isochrone['slopes'][i]*test['bp_rp_error'])**2)
        
        centers.append(center)
        tests.append(test)
        err_means.append(test['y_err'].mean())
    except:
        continue
        print(i)
    
```

# When making a standard scaler transformation the variance from the error decrease by factor of $1/s^2$

```python
for i in range(35):
    var=1#tests[i]['y'].replace('nan',0.0).fillna(0.0).astype(float).values.var()
    tests[i]['y']=scaler.fit_transform(tests[i]['y'].replace('nan',0.0).fillna(0.0).astype(float).values.reshape(-1,1))
    tests[i]['y_err']=tests[i]['y_err'].replace('nan',0.0).fillna(0.0).astype(float)/var
    gmm_GES=XDGMM(2).fit(tests[i]['y'].values.reshape(len(tests[i]),1),tests[i]['y_err'].values.reshape(len(tests[i]),1,1),R=np.array([1.0]*len(tests[i])).reshape(len(tests[i]),1,1))
    prob_ges=gmm_GES.prob_z_given_w(tests[i]['y'].values.reshape(len(tests[i]),1),tests[i]['y_err'].values.reshape(len(tests[i]),1,1),R=np.array([1.0]*len(tests[i])).reshape(len(tests[i]),1,1))
    if((prob_ges.transpose()[0]<=0.2).sum()>(prob_ges.transpose()[0]>0.2).sum()):
        tests[i]['probnew']=1-prob_ges.transpose()[0]
    else:
        tests[i]['probnew']=prob_ges.transpose()[0]
```

```python
def extinction(alpha,orig_scatter,isochrone2):
    p_slopes=[]
    orig_scatter['mean_absolute_mag_g_band_n']=orig_scatter['mean_absolute_mag_g_band']-alpha[0]
    orig_scatter['bp_rp_n']=orig_scatter['bp_rp']-alpha[1]
    x=isochrone2['BPRP']
    y=isochrone2['Gaia_G_DR2Rev']
    for i in range(len(isochrone2)-1):    
        dy=y[i+1]-y[i]
        dx=x[i+1]-x[i]
        p_slopes.append(-1*dx/dy)
    p_slopes.append(0)

    isochrone2['p_slopes']=p_slopes
    isochrone2['slopes']=-1/isochrone2['p_slopes']

    high_c=[]
    low_c=[]
    for i in range(len(isochrone2)-1):
        high_c.append(isochrone2['Gaia_G_DR2Rev'][i+1] - isochrone2['p_slopes'][i]*isochrone2['BPRP'][i+1])
        low_c.append(isochrone2['Gaia_G_DR2Rev'][i] - isochrone2['p_slopes'][i]*isochrone2['BPRP'][i])
    high_c.append(0)
    low_c.append(0)

    isochrone2['low_c']=high_c #high c is low c oops
    isochrone2['high_c']=low_c

    isochrone2=isochrone2.loc()[0:len(isochrone)-2]
    centers=[]
    tests=[]
    err_means=[]
    for i in range(len(isochrone2)):
        #try:
        x=(orig_scatter['mean_absolute_mag_g_band_n']-orig_scatter['bp_rp_n']*isochrone2['p_slopes'][i]<=isochrone2['high_c'][i])
        y=(isochrone2['low_c'][i]<=orig_scatter['mean_absolute_mag_g_band_n']-orig_scatter['bp_rp_n']*isochrone2['p_slopes'][i])
        
        test=orig_scatter[x]
        test=test[y]
        
        center=isochrone2['Gaia_G_DR2Rev'][i] - isochrone2['slopes'][i]*isochrone2['BPRP'][i]
        test['y']=test['mean_absolute_mag_g_band_n']-test['bp_rp_n']*isochrone2['slopes'][i]
        #test['y_err']=test['phot_g_mean_mag_error']**2+(isochrone2['slopes'][i]*test['bp_rp_error'])**2
        
        centers.append(center)
        tests.append(test)
        #err_means.append(test['y_err'].mean())
        #except:
        #    print("loop1")
        #    print(i)
        #    continue
            

    log_prob=0.0
    for i in range(len(isochrone2)):
        #try:
        if(len(tests[i])==0):
            log_prob+=0.0
        

        
        else:
            tests[i]['y']=tests[i]['y'].replace('nan',0.0).fillna(0.0).astype(float).values
            var=tests[i]['y'].var()
            mu=centers[i]

            #tests[i]['y']=scaler.fit_transform(tests[i]['y'].replace('nan',0.0).fillna(0.0).astype(float).values.reshape(-1,1))
            #tests[i]['y_err']=tests[i]['y_err'].replace('nan',0.0).fillna(0.0).astype(float)/var
            
            #gmm_GES=XDGMM(2).fit(tests[i]['y'].values.reshape(len(tests[i]),1),tests[i]['y_err'].values.reshape(len(tests[i]),1,1),R=np.array([1.0]*len(tests[i])).reshape(len(tests[i]),1,1))
            #log_prob+=gmm_GES.logL(tests[i]['y'].values.reshape(len(tests[i]),1),tests[i]['y_err'].values.reshape(len(tests[i]),1,1),R=np.array([1.0]*len(tests[i])).reshape(len(tests[i]),1,1)).sum()

            def gaussian(x):
                return np.log(1.0/np.sqrt(2.0*np.pi*var**2))-(np.power(x - mu, 2.) / (2 * np.power(var, 2.)))
            tests[i]['prob_g']=tests[i]['y'].apply(gaussian)
            log_prob+=np.abs(tests[i]['y']-centers[i]).sum()
            #log_prob+=(tests[i]['prob_g']).sum()
        
    
    return log_prob
```
